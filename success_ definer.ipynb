```python
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Filter the DataFrame for 'Mentor' responses
df_mentor = success_pairs_final.dropna(subset=['Response'])
df_mentor = df_mentor[df_mentor['Mentor'] == 'Mentor']

# Calculate sentiment for 'Mentor' responses
df_mentor['sentiment'] = df_mentor['Response'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Filter the DataFrame for 'Mentee' responses
df_mentee = success_pairs_final.dropna(subset=['Response'])
df_mentee = df_mentee[df_mentee['Mentor'] == 'Mentee']

# Calculate sentiment for 'Mentee' responses
df_mentee['sentiment'] = df_mentee['Response'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Concatenate the two DataFrames
df_combined = pd.concat([df_mentor, df_mentee])

# Plot the combined sentiment distribution
sns.histplot(data=df_combined, x='sentiment', hue='Mentor', kde=True)
plt.title('Sentiment Distribution of Mentor and Mentee Responses')
plt.show()
```
