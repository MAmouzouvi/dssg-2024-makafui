```python
import os
import pandas as pd
import docx
import re

def docx_to_dataframe(docx_filename):
    doc = docx.Document(docx_filename)
    full_text = "\n".join(para.text for para in doc.paragraphs)
    segments = re.split(r'(Mentor commented at|Mentee commented at|Mentee \d{4}-\d{2}-\d{2}, \d{2}:\d{2}:|Mentor \d{4}-\d{2}-\d{2}, \d{2}:\d{2}:)', full_text, flags=re.IGNORECASE)
    data = []

    for i in range(1, len(segments), 2):
        person = segments[i].strip()
        text = segments[i + 1].strip()
        data.append({'Person': person, 'Text': text})

    return pd.DataFrame(data)

def process_file(docx_filename, output_csv_filename):
    
    df = docx_to_dataframe(docx_filename)

    df['Date'] = df['Text'].str.extract(r'(\b\d{1,2}:\d{2}(?:AM|PM) [A-Z][a-z]+ \d{1,2}\b)')
    df = df.dropna()
    df = df.drop_duplicates(subset='Date')

    df.to_csv(output_csv_filename, index=False)

file_mappings = {
    'Datasets/individual_relationships/Relationship 40437.docx': 'clean_data/40437.csv',
    'Datasets/individual_relationships/Relationship 51050.docx': 'clean_data/51050.csv',
    'Datasets/individual_relationships/Relationship 55268.docx': 'clean_data/55268.csv',
    'Datasets/individual_relationships/Relationship 59809.docx': 'clean_data/59809.csv',
    'Datasets/individual_relationships/Relationship 64642.docx': 'clean_data/64642.csv',
    'Datasets/individual_relationships/Relationship 67329.docx': 'clean_data/67329.csv',
    'Datasets/individual_relationships/Relationship 59057.docx': 'clean_data/59057.csv',
}

for docx_filename, csv_filename in file_mappings.items():
    process_file(docx_filename, csv_filename)
```
